### **Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('compressed_data.csv.gz')
df
```

```python
df.info()
```

### **Handling Missing Values**

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace=True)
df.dropna(subset=['NAME', 'host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

### **Removing extra columns(which are not in used)**

```python
df = df.drop(columns=['license', 'house_rules'],errors='ignore')
df.head()
```

### **Remove dollar signs and convert to floating value**

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

### **Remove Duplicates**

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
df.describe ()
```

### **Visualization**

### **1. What is the distribution of listing prices ?**

```python
prices = np.sort(df['price'])
cdf = np.arange(len(prices)) / float(len(prices))

plt.figure(figsize=(10, 6))
plt.plot(prices, cdf, marker='.', linestyle='none', color='blue')
plt.title("Cumulative Distribution of Listing Prices")
plt.xlabel("Price (in $)")
plt.ylabel("Cumulative Proportion")
plt.grid(True)
plt.show()
```

# 🔍Insights:

* "70% of listings are priced below $250, while 90% are below $500."

* Clients can immediately see the affordability threshold in the market.

* Helps answer business questions like: "At what price point do we capture most customers?"

# 🎯 Business Takeaway

"The Airbnb price distribution is highly skewed, with the majority of listings in the budget to mid-range segment. Luxury listings exist but form a very small portion. For pricing and positioning, the client should focus primarily on the affordable segment (where demand is highest), while treating luxury properties as a niche offering that can be marketed differently."

### **4. What is the relationship between price and room type ?**

### **5. How has the number of reviews changed over time ?**

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size().reset_index(name='count')

plt.Figure(figsize=(12, 6))
reviews_over_time.plot(x='last review', y='count', kind='line', marker='o')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

# 🔍 Insights

* Growth pattern: A steady increase in reviews over time typically reflects rising platform adoption and demand.

* Seasonality: Peaks may appear during holiday or vacation seasons (e.g., summer, year-end holidays).

* Recent trends: Any sharp drop in reviews might signal external disruptions (e.g., COVID-19, local regulations, travel restrictions).

* User activity: Higher review counts = higher customer engagement → indicates trust and repeat usage.

# 🎯 Business Takeaway

"Review activity is a strong proxy for platform usage. Increasing trends suggest growing customer adoption, while seasonal peaks highlight periods of high travel demand. Any recent decline in reviews may require investigation into external factors or shifts in user behavior."

* Bronx & Staten Island: Have the lowest average prices, indicating affordability but also lower demand or fewer luxury properties.

# 🎯 Business Takeaway

"Location strongly drives pricing. Manhattan and Brooklyn command premium rates, while Queens offers mid-range options. Bronx and Staten Island remain budget-friendly markets with potential for growth, but likely attract a different customer segment."

### **7. What is the Average Availability by Room Type ?**

```python
avg_availability = df.groupby('room type')['availability 365'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_availability, x='room type', y='availability 365', palette='viridis')

plt.title('Average Availability by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Availability (days)')
plt.show()
```

# 🔍 Insights

* Entire home/apt: Often show lower average availability, since many are used as secondary homes or seasonal rentals.

* Private rooms: Tend to have higher availability, as hosts can offer them year-round while living in the same property.

* Shared rooms: Also usually available for long periods, reflecting budget accommodations with consistent access.

* Hotel rooms: Typically close to full availability (365 days), since they operate professionally.

# 🎯 Business Takeaway

"Room availability varies significantly by type. Hotel-style listings and private/shared rooms are usually available year-round, while entire homes/apartments may be restricted to seasonal or occasional rental. This suggests different revenue models: professional operators maximize occupancy, while individual hosts focus on flexibility."

### **10. What is the Revenue Potential by Neighbourhood ?**

```python
revenue_potential = (df.groupby('neighbourhood')
                        .apply(lambda x: x['price'].mean() * x['availability 365'].mean())
                        .reset_index(name='revenue_potential')
                        .sort_values(by='revenue_potential', ascending=False)
                        .head(10))

plt.figure(figsize=(12, 6))
sns.barplot(data=revenue_potential, x='neighbourhood', y='revenue_potential', palette='viridis')

plt.title('Top 10 Neighbourhoods by Revenue Potential')
plt.xlabel('Neighbourhood')
plt.ylabel('Revenue Potential')
plt.xticks(rotation=45)
plt.show()
```

# 🔍Insights:

* Certain neighbourhoods dominate due to both higher prices and higher availability.

* Top areas likely align with high-demand tourist destinations.

* Lower-ranked neighbourhoods may still be profitable but are less competitive in pricing or availability.

# 🎯Business Takeaway

“High-revenue neighbourhoods represent hotspots for investment and growth. Focusing marketing, host recruitment, or expansion efforts in these areas could maximize returns.”

* price and availability_365 often have little or no correlation → pricing doesn’t strongly depend on availability.
